```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
x_train = x_train.astype('float64')
x_test = x_test.astype('float64')
```

```python
x_train.shape
```

```python
# label the dep variable - classification problem
y_train1 = keras.utils.to_categorical(y_train)
y_test1 = keras.utils.to_categorical(y_test)
```

```python
y_train1[0]
```

```python
y_train1.shape
```

```python
y_train1[1050]
```

```python
# Building a model

model = None
model = Sequential()
model.add(Dense(128, input_dim = 784, activation='relu'))
#model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fit the model
model.fit(x_train, y_train1, validation_data=(x_test, y_test1), epochs=10, batch_size=128)
```
